# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os, json
```

```python
# Load final_tmdb csv files as dataframes and concatenate
tmdb_2000 = pd.read_csv("C:\\Users\\hamma\\Documents\\DataEnrichment\\Data-Enrichment-Project\\Data\\final_tmdb_data2000.csv.gz")
tmdb_2000.info()
tmdb_2000.head()
```

```python
tmdb_2001 = pd.read_csv("C:\\Users\\hamma\\Documents\\DataEnrichment\\Data-Enrichment-Project\\Data\\final_tmdb_data2001.csv.gz")
tmdb_2001.info()
tmdb_2001.head()
```

```python
tmdb_combined_df = pd.concat((tmdb_2000, tmdb_2001), ignore_index=True)
tmdb_combined_df.info()
tmdb_combined_df.head()
```

```python
# Save dataframe of all API calls to a .csv.gz file
tmdb_combined_df.to_csv('/Data/tmdb_results_combined.csv.gz')
```

# How many movies had at least some valid financial information (values > 0 for budget OR revenue)?

```python
# Filter dataframe so there is no 0 values in 'budget' OR 'venue' columns
valid_finance_df = tmdb_combined_df[(tmdb_combined_df['budget'] > 0) | (tmdb_combined_df['revenue'] > 0)]
valid_finance_df.info()
```

> There are 637 movies with valid financial information

# How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
valid_finance_df['certification'].value_counts().plot(kind='bar', 
                                                      color =['blue','orange',
                                                             'purple', 'cyan',
                                                             'black'])
plt.title('Movie Counts By Certifications')
plt.xlabel('Certifications')
plt.ylabel('Counts')
plt.tight_layout;
```

```python
cert_counts = valid_finance_df['certification'].value_counts()
cert_counts
```

**Movie Rating Counts Per Certification Category**
- G: 15
- PG: 35
- PG-13: 131
- R: 232
- NR: 17

# What is the average revenue per certification category?

```python
# Use float format so full values are displayed
pd.set_option('display.float_format', '${:,.2f}'.format)
```

```python
avg_rev = valid_finance_df.groupby('certification')['revenue'].mean()
print(f'Average Revenue Per Certification {avg_rev}')
```

```python
avg_rev.plot(kind='bar', color =['blue','orange', 'purple', 'cyan','black'])
plt.title('Average Movie Revenues Per Certification')
plt.xlabel('Certifications')
plt.ylabel('Revenue in Dollars')
plt.ticklabel_format(style='plain', axis='y')
plt.tight_layout;
```

# What is the average budget per certification category?

```python
avg_budg = valid_finance_df.groupby('certification')['budget'].mean()
print(f'Average Budget Per Certification {avg_budg}')
```

```python
avg_budg.plot(kind='bar', color =['blue','orange', 'purple', 'cyan','black'])
plt.title('Average Movie Budget Per Certification')
plt.xlabel('Certifications')
plt.ylabel('Budget in Dollars')
plt.ticklabel_format(style='plain', axis='y')
plt.tight_layout;
```

```python
avg_profit = avg_rev-avg_budg
print(f'Average Profit Per Certification {avg_profit}')
```

```python
avg_profit.plot(kind='bar', color =['blue','orange', 'purple', 'cyan','black'])
plt.title('Average Profit Per Certification')
plt.xlabel('Certifications')
plt.ylabel('Profit in Dollars')
plt.ticklabel_format(style='plain', axis='y')
plt.tight_layout;
```
